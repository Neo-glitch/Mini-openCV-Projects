# Mini Project 2 - Identifying Contours by Shapes

```python
import numpy as np
import cv2


# code that loads and greyscales our image
image = cv2.imread('C:\image_for_computer_vision\List_of_shapes.jpg')
grey_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Identifying Shapes', image)
cv2.waitKey()

ret, thresh = cv2.threshold(grey_img,127,255,1)

# code that extract contours from our image
contours, hierachy = cv2.findContours(thresh.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

for cnt in contours:
    # code that gets the approximate contours/polygon
    approx = cv2.approxPolyDP(cnt, 0.005*cv2.arcLength(cnt, True), True)
    
    
    # code that checks if the the shape has 3 sides( meaning a triangle)
    if len(approx) == 3:
        shape_name = 'Triangle'
        cv2.drawContours(image, [cnt], 0, (0,255,0), -1)
        
        # code that finds the centroid to place the text contaning the shape name
        # N.B: since the text starts from the bottom right of the centroid, to avoid spilling the text out of the image.
        # we minus some pixels from the x_cordinates of the centroid
        M = cv2.moments(cnt)
        cx = int(M['m10']/ M['m00'])
        cy = int(M['m01']/ M['m00'])
        cv2.putText(image, shape_name,(cx-50,cy), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255),1)
        
    # code that checks if the len of the appoximate contour is four/ or if the shape ha four sides.
    elif len(approx) == 4:
        # code that creates a crop rectangle, this will be used to diff a rect and a square
        # note that; "w" is width or x2 and "h" is the height or y2
        x,y,w,h = cv2.boundingRect(cnt)
        M = cv2.moments(cnt)
        cx = int(M['m10']/ M['m00'])
        cy = int(M['m01']/ M['m00'])
        
        # code that check to see if the 4-sided polygon is a square or rectangle
        # This code abs means if the absolute value of width - height of the shape is less than 3, then surely it will be a square.
        if abs(w-h) <=5:
            shape_name = 'Square'
            
            # code that fills the shape with a colour that we specified abnd put the text of the shape name
            # This is done using the parameter; "-1"
            cv2.drawContours(image, [cnt], 0, (0,125,255), -1)
            cv2.putText(image, shape_name, (cx-50,cy), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255, 255),1)
            
        else:
            shape_name = 'Rectangle'
            
            # code that finds centroid and place text at the center
            cv2.drawContours(image, [cnt], 0,(0,0,255), -1)
            M = cv2.moments(cnt)
            cx = int(M['m10']/ M['m00'])
            cy = int(M['m01']/ M['m00'])
            cv2.putText(image, shape_name, (cx-50, cy), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255),1)
            ,
    # code that checks if the shape has 10 sides( a star shape):
    elif len(approx) == 10:
        shape_name = 'Star'
        cv2.drawContours(image, [cnt], 0,(255,255,0), -1)
        M = cv2.moments(cnt)
        cx = int(M['m10']/ M['m00'])
        cy = int(M['m01']/ M['m00'])
        cv2.putText(image, shape_name, (cx-50, cy),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),1)
        
    # code that checks if the approximate contours is greater than or eqauls to 12 ( a circle in this case)
    elif len(approx) >= 12:
        shape_name = 'Circle'
        cv2.drawContours(image, [cnt], 0, (0,255,255), -1)
        M = cv2.moments(cnt)
        cx = int(M['m10']/ M['m00'])
        cy = int(M['m01']/ M['m00'])
        cv2.putText(image, shape_name,(cx-50,cy), cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),1)
    
    
    cv2.imshow('Identifying Shapes', image)
    cv2.waitKey()
    
cv2.destroyAllWindows()
            
```
