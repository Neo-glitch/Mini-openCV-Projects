### Mini Project - Counting Circles and Ellipses

```python
import cv2
import numpy as np

# loads and greyscales image in one line
image = cv2.imread(r'C:\Users\princ\Documents\Mini openCV projects\image_for_computer_vision\Blobs.jpg',0)
cv2.imshow('Original image', image)
cv2.waitKey(0)

# reates the detector object using the default parameters
detector = cv2.SimpleBlobDetector_create()


# detects blobs or keypoints
keypoints = detector.detect(image)

# draw blobs on our image as blue
blank = np.zeros((1,1))
blobs = cv2.drawKeypoints(image, keypoints, blank,(255,0,0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

number_of_blobs = len(keypoints)
text = 'Total Number of Blobs: ' + str(number_of_blobs)
cv2.putText(blobs, text, (20,50), cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),3)


cv2.imshow('Blobs detection using the default parameters', blobs)
cv2.waitKey()
cv2.destroyAllWindows()



# gets our filtering parameters, This is what helps us choose on which shape we want to detect.
# code that initializes parameter setting using cv2.SimpleBlobDetector_params
params = cv2.SimpleBlobDetector_Params()


# set Area filtering Parameters
params.filterByArea = True
params.minArea = 100


# sets the Circularity filtering parameters
params.filterByCircularity = True
params.minCircularity = 0.7

# sets the Convexity filtering Parameters
params.filterByConvexity = True
params.minConvexity = 0.2

# set the Inertia filtering parameters
params.filterByInertia = True
params.minInertiaRatio = 0.01


# creates the detector object after specifying the parametrs.
# NOTE: This practice is a must.
detector = cv2.SimpleBlobDetector_create(params)

# that detects the blobs
keypoints = detector.detect(image)

# code that draws green circles on the detected blobs
blank = np.zeros((1,1))
blobs = cv2.drawKeypoints(image,keypoints, blank, (0,255,0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

number_of_blobs = len(keypoints)
text = 'The Number of Circular Blobs: ' + str(number_of_blobs)
cv2.putText(blobs, text,(10,100), cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)

# code that show blobs
cv2.imshow('Circular blob detection', blobs)
cv2.waitKey()
cv2.destroyAllWindows()
```

NOTE BELOW:

a. filtering by area when using the cv2.SimpleBlobDetector_Params function;

note: Here we set True or False if we want to filter by area or not.

note: then we can set the min area that the blob detector can be able to detect blobs. and also we can set the max area if we want, but i will just leave it like that.



b. filtering by Circularity:


note: This is a measureof being circular, we can choose if we want to filter by circularity by setting it True or False.


note: And then we set the min circularity value; "1" being for a perfect circle and "0" being for the opposite.



c. filtering by Convexity:

note: This filters by measuring the area of the convex hull.


note: if can set it to True / False depending on whether we want to use it.


note: and then we set the min convexity value from "0 - 1".





d. filtering by inertia:


note: as used by our lecturer, it is a measure of ellipticalness


note: where "low inertia" being more elliptical or ovalshape and "high" being more circular.


note: we set it to True/ False depending on whether we want to use the filter or not.


note: After which we set the minInertiaRatio





NOTE: it's good practice and also a must set the parameters first before creating the simple Blob detector with our choosen paramaetrs.
